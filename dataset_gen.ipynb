```python
types = [
    "{a} - {b}",
    "{a} + {b}",
    "{a} / {b}",
    "{a} * {b}",
    "{a} + {b} - {c}",
    "{a} + {b} * {c}",
    "{a} - {b} / {c}",
    "{a} * {b} / {c}",
    "{a} * ({b} + {c})",
    "{a} * ({b} / {c})",
    "{a} + ({b} - {c})",
    "{a} * ({b} / {c})",
    "{a} / ({b} / {c})",
    "{a} * ({b} * {c})",
    "({a} - {b}) * {c}",
    "({a} - {b}) + {c}",
    "({a} - {b}) / {c}",
    "({a} - {b}) - {c}",
    "({a} * {b}) * {c}",
    "({a} / {b}) * {c}"
]
```

```python
expr = []
for type in types:
    for a in range(2,100,4):
        for b in range(2,100,4):
            if "c" in expr:
                for c in range(1,20,4):
                    if eval(type.format(a=a,b=b,c=c)) != 0:
                        expr.append(type.format(a=a,b=b,c=c))
                        print(expr[-1],end="\r")
            else:
                if eval(type.format(a=a,b=b)) != 0:
                    expr.append(type.format(a=a,b=b))
                    print(expr[-1],end="\r")
```

```python
ds = []
for exp in expr:
    d = {
        "question":{
            "type":"Text",
            "value":exp
        },
        "answers":[
            {
                "type":"Text",
                "value":(round(eval(exp),2)),
                "answer":True
            },
            {
                "type":"Text",
                "value":(eval(exp)-round(eval(exp)*.1,2)),
                
            },
            {
                "type":"Text",
                "value":(eval(exp)+round(eval(exp)*.1,2)),
                
            },

        ]
    }
    ds.append(d)
    
```

```python
with open("data.json","w+") as f:
    f.write(json.dumps(ds))
    f.close()
```
